```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")
tesla_data.reset_index(inplace=True)
print(tesla_data.head())
plt.figure(figsize=(12,6))
plt.plot(tesla_data['Date'], tesla_data['Close'], label='Close Price', color='blue')
plt.title("Historical Share Price")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.show()
tesla = yf.Ticker("TSLA")
financials = tesla.financials
revenue = financials.loc["Total Revenue"]
revenue_df = revenue.reset_index()
revenue_df.columns = ['Date', 'Revenue']
revenue_df['Date'] = pd.to_datetime(revenue_df['Date'])
revenue_df = revenue_df.sort_values('Date')
plt.figure(figsize=(12,6))
plt.plot(revenue_df['Date'], revenue_df['Revenue']/1e9, marker='o', color='orange')
plt.title("Historical Revenue")
plt.xlabel("Date")
plt.ylabel("Revenue (Billion USD)")
plt.grid(True)
plt.show()
```

```python
import yfinance as yf
import pandas as pd
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")
gme_data.reset_index(inplace=True)
gme_data.head()
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.plot(gme_data['Date'], gme_data['Close'], color='blue')
plt.title("Historical Share Price")
plt.xlabel("Date")
plt.ylabel("Close Price (USD)")
plt.grid(True)
plt.show()
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Tạo đối tượng cổ phiếu GME
gme = yf.Ticker("GME")

# Lấy báo cáo tài chính (Income Statement)
financials = gme.financials

# Lấy dòng Total Revenue
revenue = financials.loc["Total Revenue"]

# Chuyển thành DataFrame
revenue_df = revenue.reset_index()
revenue_df.columns = ['Date', 'Revenue']

# Chuyển Date sang datetime
revenue_df['Date'] = pd.to_datetime(revenue_df['Date'])

# Lọc từ năm 2004 trở đi
revenue_df = revenue_df[revenue_df['Date'].dt.year >= 2004]

# Sắp xếp theo thời gian tăng dần
revenue_df = revenue_df.sort_values('Date')

# Vẽ biểu đồ
plt.figure(figsize=(12,6))
plt.plot(revenue_df['Date'], revenue_df['Revenue']/1e9, marker='o', color='green')
plt.title("GameStop Total Revenue (2004 - Present)")
plt.xlabel("Year")
plt.ylabel("Revenue (Billion USD)")
plt.grid(True)
plt.show()
```
